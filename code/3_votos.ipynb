# Carregando a base de dados

Atenção para as colunas QT_APTOS, QT_COMPARECIMENTO, QT_ABSTENÇÔES: esses são os números da urna. Cuidado para não soma-los ao fazer algum tipo de agregação.

```python
dtype = {'ANO_ELEICAO':'str',
          'NR_TURNO':'str',
          'SG_UF':'str',
          'CD_MUNICIPIO':'str',
          'NM_MUNICIPIO':'str',
          'NR_ZONA':'str',
          'NR_SECAO':'str',
          'NR_LOCAL_VOTACAO':'str',
          'CD_CARGO_PERGUNTA':'str',
          'DS_CARGO_PERGUNTA':'str',
          'NR_PARTIDO':'str',
          'SG_PARTIDO':'str',
          'NM_PARTIDO': 'str',
          'QT_APTOS':'str',
          'QT_COMPARECIMENTO':'str',
          'QT_ABSTENCOES':'str',
          'CD_TIPO_VOTAVEL':'str',
          'DS_TIPO_VOTAVEL':'str',
          'NR_VOTAVEL':'str',
          'NM_VOTAVEL':'str',
          'QT_VOTOS':'int'}

usecols = [
    'ANO_ELEICAO', 'NR_TURNO',
    'SG_ UF', #no arquivo tem um espaço sobrando
    'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_ZONA', 'NR_SECAO', 'NR_LOCAL_VOTACAO',
    'CD_CARGO_PERGUNTA', 'DS_CARGO_PERGUNTA', 
    'NR_PARTIDO', 'SG_PARTIDO', 'NM_PARTIDO',
    'QT_APTOS', 'QT_COMPARECIMENTO', 'QT_ABSTENCOES', 
    'CD_TIPO_VOTAVEL', 'DS_TIPO_VOTAVEL', 'NR_VOTAVEL', 'NM_VOTAVEL', 'QT_VOTOS'
]
```

```python
from zipfile import ZipFile
```

```python
zip_file = ZipFile("/content/gdrive/MyDrive/dados/votos/01_boletim-urna-original/BWEB_1t_SP_101020181938.zip")
data_df = pd.read_csv(zip_file.open("bweb_1t_SP_101020182030.csv"), encoding='Latin1', sep=';', header=0, dtype=dtype, usecols=usecols)
```

```python
data_df.rename(columns={'SG_ UF': 'SG_UF'}, inplace=True)
```

```python
data_df.head(3)
```

```python
votos_geral = data_df[data_df['CD_MUNICIPIO'] == '71072']
votos_geral.reset_index(inplace=True, drop=True)
```

```python
votos_geral
```

```python
#Carregar a base de dados do botetim de urna disponibilizada pelo TSE. 
#A base tem recorte estadual
%%time
#votos_geral = pd.read_csv("/content/gdrive/MyDrive/dados/votos/01_boletim-urna-original/bweb_1t_SP_181120201549-002.csv",
                              encoding='Latin1', sep=';', header=0, 
                              dtype=dtype,
                              usecols=usecols
                             )
```

```python
#Criando o local_unico - chave que permite cruzar com a base dos locais
votos_geral['local_unico'] = votos_geral.CD_MUNICIPIO + '-' + votos_geral.NR_ZONA + '-' + votos_geral.NR_LOCAL_VOTACAO
```

```python
#Visão geral da base
votos_geral.head(3)
```

```python
#Acrescenta o Códito TSE de identificação dos municípios

#Carregar a base
correspondencia = pd.read_csv("/content/gdrive/MyDrive/dados/locais/correspondencia-tse-ibge.csv", dtype={'GEOCOD_IBGE':'str','COD_TSE':'str'})

#preenche com 5 digitos
correspondencia["COD_TSE"] = correspondencia.COD_TSE.str.zfill(5)
```

```python
#Merge da base de correspondencia
base_aux = votos_geral.merge(correspondencia[['GEOCOD_IBGE', 'COD_TSE']], 
                      left_on='CD_MUNICIPIO', right_on='COD_TSE')

base_aux.drop(columns='COD_TSE', axis=1, inplace=True)
```

```python
base_aux
```

```python
base_aux.to_csv('/content/gdrive/MyDrive/dados/votos/02_boletim-urna-ajustado/2018_boletim_ajustado_3550308', index=False, sep=";")
```

```python
# Consolidar os votos pelo local de votação, eliminando a seção eleitoral (NR_SECAO)
# Também são desconsideradas as colunas QT_APTOS, QT_COMPARECIMENTO, QT_ABSTENÇÔES:
votos_geral_consolidado = pd.pivot_table(base_aux, index=['ANO_ELEICAO', 'NR_TURNO', 'SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO',
       'NR_ZONA', 'NR_LOCAL_VOTACAO', 
       'CD_CARGO_PERGUNTA', 'DS_CARGO_PERGUNTA', 
       'NR_PARTIDO', 'SG_PARTIDO', 'NM_PARTIDO', 'CD_TIPO_VOTAVEL', 'DS_TIPO_VOTAVEL', 'NR_VOTAVEL', 'NM_VOTAVEL', 
       'local_unico'],
       values=['QT_VOTOS'], aggfunc='sum', fill_value=0).reset_index()
```

```python
votos_geral_consolidado
```

```python
#Salvar a base estadual
votos_geral_consolidado.to_csv('/content/gdrive/MyDrive/dados/votos/03_votos_consolidado/2018_votos_3550308.csv', sep=';', index=False)
```
